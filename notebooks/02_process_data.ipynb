```python
# função para derivar a duração efetiva de cada período do jogo
# e suportar o cálculo posterior dos minutos jogados por jogador
def add_match_period_durations(matches, events_df):

    # obter o último minuto registado em cada período de cada jogo
    period_max = (
        events_df
        .groupby(["match_id", "period"], as_index=False)
        .agg(period_end_minute=("minute", "max"))
    )

    # pivotar os períodos
    period_wide = (
        period_max
        .pivot(index="match_id", columns="period", values="period_end_minute")
        .reset_index()
        .rename(columns={
            1: "first_half_end_minute",
            2: "second_half_end_minute",
            3: "extra_time_first_half_end_minute",
            4: "extra_time_second_half_end_minute",
        })
    )

    # excluir o período correspondente à disputa por penáltis por não contribuir para minutos jogados
    period_wide = period_wide.drop(columns=[5], errors="ignore")

    # identificar jogos com prolongamento a partir da existência do primeiro período extra
    period_wide["has_extra_time"] = (
        period_wide.get("extra_time_first_half_end_minute").notna()
        if "extra_time_first_half_end_minute" in period_wide.columns
        else False
    )

    minute_cols = [
        "first_half_end_minute",
        "second_half_end_minute",
        "extra_time_first_half_end_minute",
        "extra_time_second_half_end_minute",
    ]

    # manter apenas as colunas que existem no dataset, garantindo robustez entre competições
    existing_cols = [col for col in minute_cols if col in period_wide.columns]

    # calcular o minuto final do jogo com base no último período disponível
    period_wide["match_end_minute"] = period_wide[existing_cols].max(axis=1)

    # enriquecer a tabela de jogos com as durações derivadas dos eventos
    matches = matches.merge(period_wide, on="match_id", how="left")

    return matches

matches = add_match_period_durations(matches, events)

# armazenar a tabela na camada de dados processados
matches.to_parquet("../data/processed/matches.parquet")
matches
```

Com o objetivo de suportar análises condicionadas pelo tempo de jogo, foi adicionada à tabela de lineups a métrica de minutos jogados por jogador em cada partida.

```python
# converter valores temporais da StatsBomb para minutos
def time_to_minutes(value):
    if pd.isna(value) or value is None:
        return np.nan

    value = str(value)

    # tratar tempos no formato "minuto:segundo"
    if ":" in value:
        m, s = value.split(":")
        return int(m) + float(s) / 60

    return float(value)


# calcular os minutos restantes desde um determinado período até ao final do jogo
def minutes_from_period_to_end(start_minute, start_period, match_row):
    periods = []

    # construir lista de períodos disponíveis para o jogo
    if not pd.isna(match_row["first_half_end_minute"]):
        periods.append((1, 0, match_row["first_half_end_minute"]))
    if not pd.isna(match_row["second_half_end_minute"]):
        periods.append((2, 45, match_row["second_half_end_minute"]))
    if not pd.isna(match_row["extra_time_first_half_end_minute"]):
        periods.append((3, 90, match_row["extra_time_first_half_end_minute"]))
    if not pd.isna(match_row["extra_time_second_half_end_minute"]):
        periods.append((4, 105, match_row["extra_time_second_half_end_minute"]))

    total = 0.0
    start_period = int(start_period)

    # acumular apenas o tempo jogado a partir do período de entrada
    for i, (period_num, period_start, period_end) in enumerate(periods):
        if period_num < start_period:
            continue

        if period_num == start_period:
            total += period_end - start_minute
        else:
            prev_end = periods[i - 1][2]
            total += period_end - prev_end

    return total


# calcular os minutos jogados por jogador a partir da informação de lineups e substituições
def calculate_minutes_played(lineups_raw, matches):
    records = []

    for _, row in lineups_raw.iterrows():

        # obter informação posicional do jogador no jogo
        positions = row.get("positions", [])
        if not isinstance(positions, list) or len(positions) == 0:
            continue

        # considerar o primeiro intervalo de utilização registado para o jogador
        envelope = positions[0]

        start_minute = time_to_minutes(envelope.get("from"))
        start_period = envelope.get("from_period")
        end_reason = envelope.get("end_reason")
        end_raw = envelope.get("to")

        # assumir início do jogo quando o minuto de entrada não está preenchido
        if pd.isna(start_minute) or start_minute is None:
            start_minute = 0.0

        # obter dados temporais do jogo correspondente
        match_row = matches[matches["match_id"] == row["match_id"]].iloc[0]

        # se o jogador termina em campo, calcular minutos até ao apito final
        if end_reason == "Final Whistle" or end_raw is None:
            minutes = minutes_from_period_to_end(start_minute, start_period, match_row)

        # se o jogador sai antes do fim, calcular diferença direta entre saída e entrada
        else:
            end_minute = time_to_minutes(end_raw)
            minutes = max(0, end_minute - start_minute)

        records.append({
            "player_id": row["player_id"],
            "player_name": row["player_name"],
            "player_nickname": row["player_nickname"],
            "jersey_number": row["jersey_number"],
            "country": row["country"],
            "team_name": row["team_name"],
            "team_id": row["team_id"],
            "match_id": row["match_id"],
            "minutes_played": minutes,
        })

    return pd.DataFrame(records)


# garantir que a coluna positions está em formato lista antes do processamento
lineups["positions"] = lineups["positions"].apply(
    lambda x: x.tolist() if isinstance(x, np.ndarray) else x
)

# calcular minutos jogados e armazenar a tabela na camada de dados processados
lineups = calculate_minutes_played(lineups, matches)
lineups.to_parquet("../data/processed/lineups.parquet")

lineups
```

# Métrica Line-Breaking Pass Value

A métrica **Line-Breaking Pass Value (LBPV)** procura quantificar o valor contextual de passes que conseguem penetrar estruturas defensivas adversárias. Em vez de medir apenas o número de passes progressivos ou o número de linhas quebradas, a métrica avalia a capacidade de um passe gerar vantagem posicional e desorganizar o bloco defensivo através da combinação entre progressão no terreno de jogo, linhas defensivas quebradas, número de adversários ultrapassados e consequência ofensiva da ação.

## Definição

Um **passe que quebra linhas** é um passe rasteiro ou baixo que penetra fisicamente um bloco defensivo e permite avançar a bola de forma significativa no terreno de jogo. Os filtros iniciais aplicados foram os seguintes:

- Apenas passes com `pass_height` igual a `"Ground Pass"` ou `"Low Pass"`
- Apenas passes concluídos com sucesso
- Exclusão de bolas paradas e cruzamentos
- Exclusão de passes realizados por guarda-redes


## Critérios Geométricos

Para que um passe seja classificado como line-breaking, tem de satisfazer **todas** 
as seguintes condições de forma simultânea:

### 1. Avanço mínimo — o passe tem de ganhar terreno (progressivo)

O passe tem de avançar pelo menos **12 metros no eixo X** (direção de ataque), 
o equivalente a 10% do comprimento total do campo StatsBomb (120m).


<div align="center">

`distance_advanced = end_x − pass_x ≥ 12m`

</div>


### 2. Linha defensiva no trajeto do passe

Deve existir no mínimo uma linha defensiva **entre o passador e o destino do passe**.

### Como são identificadas as linhas defensivas

As linhas são detectadas automaticamente a partir do **freeze frame** do 
StatsBomb 360.

Filtra-se para os adversários (excluindo guarda-redes) e agrupa-se por 
proximidade em X usando **clustering por gaps**:

> Se dois defesas consecutivos (ordenados por X) estiverem a **mais de 4,5m** 
> de distância entre si, pertencem a linhas defensivas diferentes.

Apenas grupos com **2 ou mais defesas** são considerados uma linha. 
Jogadores isolados não formam uma linha e são ignorados na deteção.

Para cada linha, calcula-se a posição média em X dos defesas que a formam 
(`line_x`). Essa posição tem de estar contida no intervalo:

<div align="center">

`pass_x ≤ line_x ≤ end_x`

</div>


### 3. O passe tem de passar entre dois defesas
Não basta o passe avançar além de uma linha, este tem de passar **pelo espaço entre 
dois defesas adjacentes** dessa linha.
Para isso, calcula-se onde a trajetória do passe está em `line_x` 
(interpolação linear) e verifica-se se esse ponto Y cai entre dois defesas 
consecutivos ordenados por Y:

<div align="center">

`y_defesa_i ≤ y_trajeto ≤ y_defesa_i+1`

</div>

Esta restrição foi introduzida para evitar que passes que contornam a linha defensiva pela largura (por fora do último defesa) sejam classificados como line-breaking, uma vez que essas ações exploram espaço lateral em vez de romper diretamente a estrutura defensiva adversária.


### 4. Distância mínima ao ponto de cruzamento

A distância Euclidiana entre o passador e o ponto onde a trajetória do passe cruza a 
linha defensiva tem de ser **≥ 5 metros**:

<div align="center">

`√((line_x − pass_x)² + (y_crossing − pass_y)²) ≥ 5m`

</div>


Esta condição evita falsos positivos em situações onde o passador está já 
encostado à linha, e um passe parece geometricamente 
"romper" a linha quando na prática não há qualquer disrupção tática.

## Metric Score

O score final da métrica é calculado através de uma combinação ponderada de diferentes componentes contextuais associadas ao passe. Cada componente é normalizada para o intervalo `[0,1]`, permitindo comparar ações com características distintas numa escala comum.

<div align="center">

`score = Σ (peso × componente_normalizada)`

</div>

| Componente | Peso | O que representa |
|---|---:|---|
| `zone_value` | 10% | Valor posicional da zona onde o passe tem origem |
| `distance_norm` | 15% | Capacidade de progressão territorial do passe |
| `defenders_norm` | 30% | Número de adversários ultrapassados entre origem e destino |
| `line_break_norm` | 20% | Número de linhas defensivas efetivamente quebradas |
| `outcome_norm` | 25% | Consequência ofensiva gerada após o passe |

#### `zone_value`
Mede o valor posicional da origem do passe através de uma função sigmoide baseada na coordenada `x` do passe. A função atribui maior valor a passes iniciados em zonas mais avançadas do terreno, com crescimento mais acentuado a partir de zonas intermédias/ofensivas.

#### `distance_norm`
Representa a progressão territorial do passe no eixo X, calculada através de `end_x - pass_x`. A variável é normalizada relativamente ao maior valor observado nos passes identificados como line-breaking.

#### `defenders_norm`
Quantifica o número de adversários cuja coordenada X se encontra estritamente entre a origem e o destino do passe. Funciona como aproximação ao número de jogadores ultrapassados pela ação. O valor é normalizado relativamente ao máximo observado no conjunto de passes identificados.

#### `line_break_norm`
Representa o número de linhas defensivas quebradas pelo passe. O valor é normalizado dividindo o número de linhas quebradas por `3`, com limite máximo de `1.0`.

#### `outcome_norm`
Avalia a consequência ofensiva da ação através de remates gerados nos `10` segundos seguintes dentro da mesma posse. Quando o passe é identificado como assistência direta para remate ou golo (`pass_shot_assist = True` ou `pass_goal_assist = True`), recebe crédito total do xG da finalização. Caso exista remate dentro da janela temporal mas sem assistência direta identificada, o valor é penalizado por um fator de `0.6`, representando uma contribuição ofensiva indireta. Se o remate resultar em golo, o valor base é `1.0`, também ajustado pelo mesmo critério de diretividade.

A soma dos pesos é igual a `1.0`, pelo que o score final fica naturalmente limitado ao intervalo `[0,1]`.

## Limitações

**Freeze frames estáticos** — A deteção de linhas defensivas é baseada exclusivamente no freeze frame disponibilizado pelo StatsBomb 360 no instante do passe. Como apenas existe uma fotografia estática do momento, não é possível observar a movimentação contínua dos jogadores e da bola durante a trajetória do passe. Além disso, apenas os jogadores capturados no freeze frame estão disponíveis para análise, podendo existir adversários fora da área observada que influenciem a estrutura defensiva real. Com dados completos de tracking seria possível modelar de forma significativamente mais rigorosa a interação entre a trajetória da bola e o bloco defensivo, verificando em tempo real se o passe atravessou efetivamente uma linha defensiva.

**Deteção aproximada de linhas defensivas** — As linhas são inferidas através de clustering baseado apenas na coordenada X dos adversários. Embora esta abordagem permita identificar estruturas defensivas de forma simples e interpretável linhas muito compactas ou posicionamentos diagonais podem originar agrupamentos imperfeitos.

**Normalização dependente do dataset** — As componentes `distance_norm` e `defenders_norm` são normalizadas relativamente aos valores máximos presentes no dataset analisado. Como consequência, os scores devem ser interpretados sobretudo de forma contextual dentro do mesmo ambiente competitivo, sendo menos adequados para comparação direta entre competições, épocas ou datasets distintos.

```python
# VARIÁVEIS CONSTANTES
PITCH_LENGTH = 120.0
PITCH_WIDTH  = 80.0

MIN_ADVANCE_X      = 12.0   # avanço mínimo de 10% do comprimento do campo (120m)
LINE_GAP_THRESHOLD = 4.5    # gap em X > 4.5m → nova linha defensiva
MIN_LINE_SIZE      = 2      # uma linha necessita de pelo menos 2 defesas
OUTCOME_WINDOW_SEC = 10.0   # segundos para procurar remate/golo após o passe
SHOT_ASSIST_FACTOR = 0.6    # multiplicador para remates sem assistência direta
MIN_START_DIST = 5.0        # o passador deve estar a pelo menos 5m do ponto de interseção

# Pesos das componentes do score
WEIGHTS = {
    "zone":        0.10,
    "distance":    0.15,
    "defenders":   0.30,
    "line_break":  0.20,
    "outcome":     0.25,
}
```

```python
# main pipeline da métrica Line-Breaking Pass Value (LBPV)
def compute_line_breaking_passes(events: pd.DataFrame) -> pd.DataFrame:
    
    # filtrar passes progressivos e excluir guarda redes
    passes_actor = events[
        (events["type"] == "Pass") &
        (events["actor"] == True) &
        (~events["keeper"].eq(True))
    ].copy()

    # excluir bolas paradas
    setpiece_types = {"Corner", "Free Kick", "Throw-in", "Kick Off"}
    if "pass_type" in passes_actor.columns:
        passes_actor = passes_actor[
            ~passes_actor["pass_type"].isin(setpiece_types)
        ]

    # excluis passes incompletos
    non_complete = {"Incomplete", "Out", "Pass Offside", "Unknown", "Injury Clearance"}
    if "pass_outcome" in passes_actor.columns:
        passes_actor = passes_actor[
            ~passes_actor["pass_outcome"].isin(non_complete)
        ]

    # excluir cruzamentos
    if "pass_cross" in passes_actor.columns:
        passes_actor = passes_actor[
            passes_actor["pass_cross"].isna()
        ]

    # Apenas incluir passes rasteiros ou baixos
    if "pass_height" in passes_actor.columns:
        passes_actor = passes_actor[
            passes_actor["pass_height"].isin({"Ground Pass", "Low Pass"})
        ]

    # extrair coordenadas de evento
    passes_actor[["pass_x", "pass_y"]] = passes_actor["location_x"].apply(
        lambda v: pd.Series(_extract_xy(v))
    )

    # extrair coordenadas de destino do passe
    passes_actor[["end_x", "end_y"]] = passes_actor["pass_end_location"].apply(
        lambda v: pd.Series(_extract_xy(v))
    )

    # remove passes com coordenadas inválidas
    passes_actor = passes_actor.dropna(subset=["pass_x", "end_x"])

    # calcular progressão do passe no eixo X e filtrar passes progressivos
    passes_actor["distance_advanced"] = passes_actor["end_x"] - passes_actor["pass_x"]
    progressive = passes_actor[
        passes_actor["distance_advanced"] >= MIN_ADVANCE_X
    ].copy()

    # extrair linhas dos adversários presentes no freeze frame
    defenders_df = events[
        (events["type"] == "Pass") &
        (events["teammate"] == False) &
        (events["actor"] == False) &
        (~events["keeper"].eq(True))
    ][["id", "location_y"]].copy()

    # extrair coordenadas (x, y) dos adversários
    defenders_df[["def_x", "def_y"]] = defenders_df["location_y"].apply(
        lambda v: pd.Series(_extract_xy(v))
    )

    # remover adversários sem coordenadas válidas
    defenders_df = defenders_df.dropna(subset=["def_x", "def_y"])

    # agrupar coordenadas dos adversários por evento
    def_positions = (
        defenders_df
        .groupby("id")
        .apply(lambda g: list(zip(g["def_x"], g["def_y"])))
        .reset_index()
        .rename(columns={0: "defenders_xy"})
    )

    # juntar df de passes progressivos com o posicionamento dos adversários por evento
    progressive = progressive.merge(def_positions, on="id", how="left")

    # substituir valores ausentes por listas vazias
    progressive["defenders_xy"] = progressive["defenders_xy"].apply(
        lambda v: v if isinstance(v, list) else []
    )

    # calcular número de linhas quebradas e defesas ultrapassados
    results = progressive.apply(
        lambda row: pd.Series({
            "lines_broken": _count_lines_broken(
                row["pass_x"], row["pass_y"],
                row["end_x"], row["end_y"],
                row["defenders_xy"]
            ),
            "n_defenders_bypassed": _count_defenders_bypassed(
                row["pass_x"], row["end_x"],
                row["defenders_xy"]
            ),
        }),
        axis=1
    )
    progressive = pd.concat([progressive, results], axis=1)

    # filtrar os passes que quebram uma ou mais linhas
    line_breakers = progressive[progressive["lines_broken"] >= 1].copy()

    # calcular timestamps em segundos para os eventos relevantes (verificar outcome nos 10s seguintes)
    all_ts = events[[
        "id", "match_id", "possession", "type",
        "timestamp", "shot_statsbomb_xg", "shot_outcome", "actor"
    ]].copy()

    # manter apenas uma linha por evento, correspondente ao jogador que realizou a ação
    all_ts = all_ts[all_ts["actor"] == True]

    # remover coluna auxiliar depois da filtragem
    all_ts = all_ts.drop(columns=["actor"])

    # converter timestamp para segundos, facilitando o cálculo da janela temporal
    all_ts["ts_sec"] = all_ts["timestamp"].apply(_timestamp_to_seconds)

    # criar dicionário com os remates por jogo e posse
    # chave: (match_id, possession)
    # valor: lista de tuplos (timestamp do remate, xG, outcome do remate)
    shots_lookup = (
        all_ts[all_ts["type"] == "Shot"]
        .groupby(["match_id", "possession"])
        .apply(lambda g: list(zip(
            g["ts_sec"],
            g["shot_statsbomb_xg"],
            g["shot_outcome"]
        )))
        .to_dict()
    )

    # juntar timestamp do passe aos passes line-breaking
    line_breakers = line_breakers.merge(
        all_ts[["id", "ts_sec"]].rename(columns={"ts_sec": "pass_ts_sec"}),
        on="id",
        how="left"
    )

    # calcular valor ofensivo gerado após cada passe
    def _get_outcome(row):

        # identificar jogo e posse do passe
        key = (row["match_id"], row["possession"])

        # obter remates dessa posse
        shots = shots_lookup.get(key, [])

        # timestamp do passe em segundos
        pass_ts = row.get("pass_ts_sec", np.nan)

        # se o passe não tiver timestamp válido, não recebe crédito ofensivo
        if np.isnan(pass_ts):
            return 0.0

        # verificar se o passe foi assistência direta para remate ou golo
        is_direct = bool(
            row.get("pass_shot_assist", False) or
            row.get("pass_goal_assist", False)
        )

        # procurar remates dentro da janela temporal definida após o passe
        for shot_ts, xg, outcome_name in shots:

            if 0 <= (shot_ts - pass_ts) <= OUTCOME_WINDOW_SEC:

                # se o remate for golo, atribuir valor máximo
                # caso não seja assistência direta, aplicar penalização
                if str(outcome_name) == "Goal":
                    return 1.0 if is_direct else 1.0 * SHOT_ASSIST_FACTOR

                # obter valor de xG; se estiver ausente, usar valor residual por defeito
                xg_val = float(xg) if xg is not None and not (
                    isinstance(xg, float) and np.isnan(xg)
                ) else 0.05

                # assistência direta recebe xG completo; contribuição indireta recebe xG penalizado
                return xg_val if is_direct else xg_val * SHOT_ASSIST_FACTOR

        # sem remate na janela temporal, não há crédito ofensivo
        return 0.0

    # aplicar cálculo de outcome ofensivo a cada passe line-breaking
    line_breakers["outcome_value"] = line_breakers.apply(_get_outcome, axis=1)

    # calcular valor posicional da origem do passe
    line_breakers["zone_value"] = line_breakers["pass_x"].apply(_zone_value)

    # normalizar progressão do passe pelo máximo observado no dataset
    max_dist = line_breakers["distance_advanced"].max()
    line_breakers["distance_norm"] = (
        line_breakers["distance_advanced"] / max(max_dist, 1.0)
    )

    # normalizar número de adversários ultrapassados pelo máximo observado no dataset
    max_def = line_breakers["n_defenders_bypassed"].max()
    line_breakers["defenders_norm"] = (
        line_breakers["n_defenders_bypassed"] / max(max_def, 1.0)
    )

    # normalizar linhas quebradas, assumindo 3 linhas como referência máxima razoável
    line_breakers["line_break_norm"] = (
        line_breakers["lines_broken"] / 3.0
    ).clip(upper=1.0)

    # limitar outcome ofensivo ao intervalo [0, 1]
    line_breakers["outcome_norm"] = line_breakers["outcome_value"].clip(upper=1.0)

    # calcular score composto ponderado
    line_breakers["score"] = (
        WEIGHTS["zone"]       * line_breakers["zone_value"] +
        WEIGHTS["distance"]   * line_breakers["distance_norm"] +
        WEIGHTS["defenders"]  * line_breakers["defenders_norm"] +
        WEIGHTS["line_break"] * line_breakers["line_break_norm"] +
        WEIGHTS["outcome"]    * line_breakers["outcome_norm"]
    )

    # selecionar colunas finais a devolver no output
    output_cols = [
        "id", "match_id", "player", "player_id", "team", "team_id",
        "period", "minute", "second", "timestamp",
        "pass_x", "pass_y", "end_x", "end_y",
        "pass_length", "pass_angle",
        "distance_advanced",
        "n_defenders_bypassed", "lines_broken",
        "zone_value", "distance_norm", "defenders_norm",
        "line_break_norm", "outcome_value", "outcome_norm",
        "score_raw", "score",
        "pass_outcome", "pass_body_part", "pass_recipient"
    ]

    # manter apenas colunas existentes no DataFrame
    output_cols = [c for c in output_cols if c in line_breakers.columns]

    # criar DataFrame final da métrica
    result = line_breakers[output_cols].copy()

    return result
```
